```python
import requests
import datetime
```

```python

schedule_url = f"http://data.nba.com/data/10s/v2015/json/mobile_teams/nba/2020/league/00_full_schedule.json"
data = requests.get(schedule_url).json()
```

```python
game_id = list(map(lambda i: i['gid'], data['lscd'][0]['mscd']['g']))
```

```python
jan_games = list(filter(
        lambda i: 
        datetime.datetime.strptime(i['gdte'], '%Y-%m-%d') <datetime.datetime(2021, 2, 1),
        data['lscd'][0]['mscd']['g']))
```
