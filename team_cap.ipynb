```python
url = 'https://www.spotrac.com/nba/cap/2020/'
```

```python
df = pd.read_html(url)
df[0].head(20)
```

```python
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager

# setup chrome browser to run splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser("chrome", **executable_path, headless=False)

browser.visit(url)
```

```python
table = browser.find_by_tag('tbody')
```

```python
table.text
```

```python
year = browser.find_by_name('year')
year.text
```

```python
for i in year.find_by_text('2022-2023'):
    print(i)
```
